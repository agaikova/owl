```python
import pandas
import ctn_benchmark
import seaborn as sns

%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = ctn_benchmark.Data('data')
df = pandas.DataFrame(data.data)
df['ratio'] = df['_childhood'] / df['_t_glasses_on']
df = df.query ('shift != -1')
df = df.query ('shift != -2')
df = df.query ('shift != -3')
df = df.query ('shift != -4')
```
